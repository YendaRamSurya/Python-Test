```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\Users\ramsu\ltphd\banknifty_feb_future_5mins.csv") 
data.head(3)
'''     data["Close"]
        data.Close
        
     are the methds to access the indexs '''
```

```python
def Bbands(instrument,n):
    
# let us calculate the Moving average 

    Ma = instrument["Close"].rolling(window =n).mean()
    
    Sd = instrument["Close"].rolling(window =n).std()
    
    instrument["lower_BB"] = Ma - (2* Sd)
    instrument["upper_BB"] = Ma + (2* Sd)
    instrument["centre_line"] = Ma
    
    return instrument
```

```python
bank_nifty = Bbands(data,22)

bank_nifty.tail()
```

```python
%matplotlib inline

plt.figure(figsize = (14,20))
plt.ylabel('price')
plt.plot(bank_nifty.lower_BB)
plt.plot(bank_nifty.centre_line)
plt.plot(bank_nifty.upper_BB)
plt.grid(True)
plt.show()
```

# what is Lambda 

The lambda operator is a way to create small anonymous functions i.e. functions without a name.

They are temporary functions i.e. they are needed only where they have been created.

The lambda feature was added in Python due to high demand from the Lisp programmers (Lisp is a programming language).

`For Example` : *Refer Below*

```python
sum = lambda x,y : x+y 

sum(10,20)
```

```python
product = lambda a,b,c,d : a*b +c -d

product(10,20,9,7)

"""Performing the pemdas"""
```

# map ()
One of the advantages of using a lambda is the map() function.

 map (lambda, sequence of lists)
map() applies the lambda function to all elements within the sequence. These elements are generally lists.

`Operation on lists using lambda is a valid operation on our DataFrame `

```python
x= [10,20,30,60]
y = [20,99,90,78,90]

# trail on unequal lists 

z=list(map(lambda x,y : x+y , x,y))
print(z)
```

```python
list(map(lambda a,b : a*b ,x,y)) 
```

# filter ()
 Another advantage of using a lambda is the filter() function. simple excel operation

 filter (lambda, list)
**It is an elegant way to filter out the required elements from a list.**


```python
list (filter(lambda x :x>20 ,x))
```

```python
signals = ['Buy','Sell','Sell','Buy','Buy','Sell','Buy'] # This is a list

list(filter(lambda x: x=="Buy",signals))
```
