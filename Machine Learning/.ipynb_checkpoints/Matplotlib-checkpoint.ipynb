```python
x = [1,2,3,4]

y = [11,22,33,44]

plt.plot(x,y);
```

```python
# 1st method
%matplotlib
fig = plt.figure()

ax = fig.add_subplot()

plt.show()
```

```python
# second method 

fig =plt.figure()

ax = fig.add_axes([1,1,1,1])

ax.plot(x,y)

plt.show();
```

```python
# 3rd method (recommended) 

fig,ax = plt.subplots()
ax.plot(x,y)
```

### Anatomy of matplot figure 

## matplot lib exmaple work flow

```python
subplots , ax = plt.subplots(figsize =(5,10))

ax.plot(x,y)

# customise the plot 

ax.set(title = "simple plot",xlabel="X-axis",ylabel ="Y-axis")

# save the whole figure 

fig.savefig("sample_plot.png")
```

## making figures with numpy arrays 



```python
# create some data 

x= np.linspace(0,10,100)
```

```python
fig,ax = plt.subplots()

ax.plot(x,x**2)
```

```python
fig,ax = plt.subplots()
ax.scatter(x,x**3);
```

```python
fig,ax = plt.subplots()
ax.scatter(x,np.sin(x))
```

```python
# making a plot from dictionary 

nutbutter_prices ={'Almond butter':10 ,'peanut butter':8 ,'cashew butter' : 20}
fig,ax =plt.subplots()
ax.scatter(nutbutter_prices.keys(),nutbutter_prices.values())
```

```python
fig,ax =plt.subplots()
ax.bar(nutbutter_prices.keys(),nutbutter_prices.values())
ax.set(title ="cost of Butter",xlabel = "this is x axis",ylabel='ths is y shit');
```

```python
#creating a histogram
fig,ax = plt.subplots(figsize =(5,10))
ax.barh(list(nutbutter_prices.keys()),list(nutbutter_prices.values()))
# ax.barh(**nutbutter_prices)
```

```python
x = np.random.randn(1000)

fig,ax = plt.subplots()
ax.hist(x);
```

```python
# sub plot option 1 

fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows =2,ncols=2,figsize =(10,5))

ax1.plot(x,x/2)
ax2.bar(nutbutter_prices.keys(),nutbutter_prices.values())
ax3.hist(x)
ax4.scatter(x,np.sin(x))
```

```python
# sub plot option 2 
fig ,ax = plt.subplots(nrows =2,ncols=2,figsize =(10,5))

ax[0,0].plot(x,x/2)
ax[0,1].bar(nutbutter_prices.keys(),nutbutter_prices.values())
ax[1,0].hist(x)
ax[1,1].scatter(x,np.sin(x))
```
