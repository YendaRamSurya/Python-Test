### Question 1


- Finding the factorial for any number using recurrence in python 


```python
def Factorial(n):
    if n ==0:
        return 1
    return n* Factorial(n-1)
print("Hence factorial of 4 is",Factorial(4),"\n","*"*100)

# Factorial for range of numbers 

for i in range(1,10):
    print(f"Hence factorial of {i} is",Factorial(i))
print("\n","*"*100)
```

### Question 2
>Q2: Write a Python program which iterates through a sequence of 50 randomly generated integers between 100 and 200 and creates two different objects, one which are even multiples of 3 and which are odd multiples of 5

```python
import numpy as np

list1 = np.random.randint(100,200,size=(1,50))
np.random.seed(10)
print(list1)

# Method without changing the size of array 

list2 = np.where(np.logical_and( np.remainder(list1,3) == 0,np.remainder(list1,2)==0),list1,'')
list3 =np.where(np.remainder(list1,5)==0,list1,'')
print("list with even numbers divisible by 3 = ",list2,"\n","*"*100)
print("list with multiples of 5 in list1 = ",list3,"\n","*"*100)

# Forming objects by changing the size of the array 

list4 =list1[0,np.all(np.logical_and( np.remainder(list1,3) == 0,np.remainder(list1,2)==0),axis=0)]
print("list with even numbers divisible by 3 = ",list4,"\n","*"*100)
list5 = list1[0,np.all(np.remainder(list1,5)==0,axis=0)]
print("list with multiples of 5 in list1 = ",list5,"\n","*"*100)
```

### Question 3
> Write a Python program to generate a sequence of 1000 numbers drawn from a standard uniform distribution and plot it on a chart. Compute its mean. Use the np.where clause to create a sequence of only numbers greater than 0.75 from the list. Plot a chart of those numbers. Check how many such numbers exist. Is it on expected lines? Comment. Please use np.random.seed(1234) at the beginning of your code.

```python

import matplotlib.pyplot as plt

np.random.seed(1234)

uni=np.random.uniform(0,1,1000)

plt.figure(figsize=(15,10))
plt.plot(uni)
plt.show()

print(f"Mean of standard uniform distribution is {np.mean(uni)})")
```

```python
uni1=np.where(uni>0.75,uni,0)
plt.figure(figsize=(10,8))
uni2 =np.where(np.logical_and(uni>0.5,uni<=0.75))
print("Numbers greater than 0.75 are :",np.count_nonzero(uni1),'\n',"probality of occurence of (0.75,1]",np.count_nonzero(uni1)/np.count_nonzero(uni),"\n","*"*100)
print("Numbers greater than 0.5 and less than 0.75 are :",np.count_nonzero(uni2),'\n',"probality of occurence of (0.5,0.75]",np.count_nonzero(uni2)/np.count_nonzero(uni),"\n","*"*100)
plt.plot(uni1)
```

```python
import pandas as pd
import pandas_datareader as data
import datetime as dt
import os
start = dt.datetime(2010,1,1)
end = dt.datetime(2020,5,1)
dfn = data.get_data_yahoo("TSLA",start,end)
os.getcwd()
#df.to_csv("TSLA.csv")
df = dfn.copy()
print(df.shape,f"\n number of rows is {df.shape[0]}",f"\n number of columns is {df.shape[1]}")

print("*"*100,'\n',"following is the summary of statistics ","\n",df.describe(),"\n","*"*100)

print(f"The Highest price reched by Tsla stock from {start} to {end} is :", df.High.max(),"\n","*"*100)
```

### Question 5
   > Q5: For the TSLA data downloaded for Q4,<br>
    (i)	Fetch all the data from October 2017 using the .loc[] and save it in a new DataFrame called df1. 'From df1, display only the rows for which the value in the Direction column is up.<br>
     (ii)	Add a new column to df1 containing the daily range (Close - Open) using a vectorized operation. <br>
    (iii)	Add another column to df1 containing the 3-day moving average of the Adj Close prices.<br>
    (iv)	Use the appropriate method to delete all the rows containing NaN values from df1.<br>


```python
# %%timeit
def function(x ,y): 
    return  "Up" if x > y  else "Down" if x < y else "Sideways"

df["prevclose"] = df.Close.shift(periods=1)
df["direction"] = df.apply(lambda x: function (x.Close,x.prevclose),axis=1)
print(df.tail(10),"\n","*"*100)
direction_group = df.groupby("direction")
df1=direction_group.get_group("Up")
print(df1,"\n","*"*100)
```

```python
# making a copy of data from Oct 2017 

df1=df1.loc["2017-10-1":,].copy()

# Making a coumn of daily range

df1["daily_range"] = df1.Close- df1.Open
print(df1.head(),"\n","*"*100)

# creating a column of 3 day moving average 

df1["3-day MA"]= df1["Adj Close"].rolling(window=3).mean() 
print(df1.head(8),"\n","*"*100)

# Dropping the na values of 3 day ema
df1=df1.dropna(axis=0,how='any')
print(df1.head(10),"\n","*"*100)

# rearranging the columns 
pd.DataFrame(df1,columns=["High","Low","Open","Close","Adj Close","direction","daily_range" ,"3-day MA","Volume"])
```
